```
import pandas as pd
import requests
import time
from datetime import datetime
from pyspark.sql.functions import *
```

```
df_weather_spark = spark.sql(f"""SELECT * FROM delta. `/dbfs/Silver/ExtraçãoTempo` """)
```

```
df_weather_spark.limit(10).display()
```

```
df_precipitation_spark = df_weather_spark.drop('evapotranspiration','temperature','humidity')
```

```
df_precipitation_spark = df_precipitation_spark.orderBy('date','time_str')
```

```
df_precipitation_spark.display()
```

```
df_precipitation_spark.repartition(20)\
              .write.format("delta")\
              .mode("overwrite")\
              .option("overwriteSchema", "true")\
              .save("/dbfs/Silver/Precipitation")
```
